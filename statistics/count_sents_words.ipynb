```python
import boto3
import os
import nltk
from tqdm import tqdm_notebook
```

```python
file_form = 'v2_chunked_data/next_contentcontentHOO{0}content.txt'
```

```python
paper_cnt = 0  #  
sent_cnt = 0  #  
words_per_sent = {}  #    
text_all = ''

for i in range(12786):
    try:
        # S3   
        file_name = file_form.format(i)
            
        #   .
        with open(file_name, 'r', encoding='UTF-8') as f:
            text = f.read()
            
        #   .
        paper_cnt += 1
        
        #   .
        sents = nltk.sent_tokenize(text)
        sent_cnt += len(sents)
        
        #     .
        for sent in sents:
            words = nltk.word_tokenize(sent)
            try:  #   
                words_per_sent[len(words)] += 1
            except KeyError:  #   
                words_per_sent[len(words)] = 1
        
    except Exception as e:
        print(i, e)
```

```python
print('  :', paper_cnt)
print('  :', sent_cnt)

#     
one_to_20 = 0
for i in range(1, 20):
    one_to_20 += words_per_sent[i]
    print(i, words_per_sent[i] / sent_cnt)
percent_20_or_more = (sent_cnt - one_to_20) / sent_cnt
print(percent_20_or_more)
```
