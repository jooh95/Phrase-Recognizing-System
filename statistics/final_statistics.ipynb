```python
import boto3
import os
import nltk
from tqdm import tqdm_notebook
```

```python
bucket = boto3.resource('s3').Bucket('learningdatajchswm9')
file_form = 'txtData/HOO{0}.txt'

if not os.path.exists('txtData'):
    os.makedirs('txtData')
```

```python
paper_cnt = 0  #  
sent_cnt = 0  #  
words_per_sent = {}  #    
text_all = ''

for i in tqdm_notebook(range(12786)):
    try:
        # S3   
        file_name = file_form.format(i)
        
        #     
        if not os.path.exists(file_name):
            bucket.download_file(file_name, file_name)
            
        #   .
        with open(file_name, 'r', encoding='UTF-8') as f:
            text = f.read()
        
        #   
        text_all += text
            
        #   .
        paper_cnt += 1
        
        #   .
        sents = nltk.sent_tokenize(text)
        sent_cnt += len(sents)
        
        #     .
        for sent in sents:
            words = nltk.word_tokenize(sent)
            try:  #   
                words_per_sent[len(words)] += 1
            except KeyError:  #   
                words_per_sent[len(words)] = 1
        
    except Exception as e:
        print(i, e)
        
print('  :', paper_cnt)
print('  :', sent_cnt)
```

```python
one_to_20 = 0
for i in range(1, 20):
    one_to_20 += words_per_sent[i]
    print(i, words_per_sent[i] / sent_cnt)
# 20  
percent_20_or_more = (sent_cnt - one_to_20) / sent_cnt
percent_20_or_more
```
